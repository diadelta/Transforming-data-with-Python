```python
import pandas as pd
```

```python
df = pd.read_csv("hn_stories.csv")
```

```python
df.columns = ["submission_time", "upvotes", "url", "headline"]
df
```

```python
headlines = df["headline"]
s = ""
for hd in headlines:
    s += str(hd).lower() + " "
s = s.split()
s
```

```python
from collections import Counter 
c = Counter()
for w in s:
    c[w] += 1
c.most_common(100)
```

```python
domains = df["url"].value_counts()[0:100]
domains
```

```python
for name, row in domains.items():
    print("{0}: {1}".format(name, row))
```

```python
import time
current = time.gmtime()
print(current)
print(current.tm_year)
print(current.tm_hour)
```

```python
from dateutil.parser import parse
import datetime
def trans(st):
    date = parse(str(st))
    return date.hour
```

```python
df["submission_hours"] = df["submission_time"].apply(trans)
df["submission_hours"].value_counts()
```
